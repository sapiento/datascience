```python
import numpy as np
import pandas as pd
```

```python
#how to read your files
df = pd.read_csv('sample_pivot.csv')
df.head()#return the first five rows of your data, but you can specify inside the blacket to return the number you want eg(def.head(10))
```

```python
df['Units'] = df['Units'].fillna(df['Units'].dropna().mean())
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.set_index(['Region','Type'],inplace=True)
df
```

```python
df.loc['North']
```

```python
df.loc['North'] .loc ['Men\'s Clothing']
df
```

```python
df.reset_index(inplace=True)
df
```

```python
#CONDITIONAL SELECTIONS using & (and) |(or)
df[(df['Region']=='South') & (df['Type']=='Women\'s Clothing') & (df['Sales'] > 500)]
```

```python
#to sort some values
#if ascending is true it will start from the minimum and if its false it will start from the maximum
df.sort_values('Units',ascending=False,inplace=True)
df
```

```python
#INFO
df.info()#it gives you all information about your data
```

```python
#DESCRIBE
df.describe()#it performs  descriptive and statistics analysis only on numerical column present
```

```python
#TRANSPOSE
df.transpose()#it turns row into column and vice versa
```

```python
df.groupby('Region').apply(max)#it is use for to apply an aggregate function on the table
```

```python
#how to create a pivot table
df_pv=pd.pivot_table(df,values=['Units','Sales'],columns='Type',index='Region',aggfunc='sum')#on default the aggrfunc is the mean
df_pv
```

```python
#MARGINS; sum all the input for rows and columns
df_pv=pd.pivot_table(df,values=['Units','Sales'],columns='Type',index='Region',aggfunc='sum',margins=True)
df_pv
```

```python
#margins_name is to give a specific name to the margin
df_pv=pd.pivot_table(df,values=['Units','Sales'],columns='Type',index='Region',aggfunc='sum',margins=True,margins_name='OVERALLTOTAL')
df_pv
```

```python
#CROSS TABULATION
pd.crosstab(df['Type'],df['Region'],margins=True,margins_name='OVERALL')
```

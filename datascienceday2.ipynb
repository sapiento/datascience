```python
import numpy as np
import pandas as pd
```

```python
columnsname = ['1st CA','2nd CA','3rd CA']
rowdata= np.arange(1,31)
np.random.seed(5)
readings = np.random.randint(3,10,90).reshape(30,3)


scores = pd.DataFrame(data= readings, index = rowdata, columns = columnsname)
scores
```

```python
scores.insert(0,'Age',np.random.randint(13,16,30).reshape(30,1))
scores
```

```python
scores.insert(4,'Examscore',np.random.randint(30,70,30).reshape(30,1))
scores
```

```python
scores
```

```python
#how to create new cloumn and new row
scores['TotalSCcore'] = scores['1st CA'] + scores['2nd CA'] + scores['3rd CA'] + scores['Examscore']
scores
```

```python
#ROW
scores.loc['Average'] = (sum(scores['Age']/len(scores)), sum(scores['1st CA']/len(scores)),sum(scores['2nd CA']/len(scores)),sum(scores['3rd CA']/len(scores)),sum(scores['Examscore']/len(scores)),sum(scores['TotalSCcore']/len(scores)))
scores
```

```python
#how to drop a column inplace=True makes it permanently while false drops it temporaly
scores.drop('TotalSCcore',axis = 1, inplace=True)
```

```python
#to drop rows
scores.drop('Average',axis = 0, inplace=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

DISTRIBUTION PLOT

```python
sns.displot(df['total_bill'],kde = True)# kernel density
```

REGRESSION PLOT

```python
sns.regplot(x= df['total_bill'], y = df['tip'])
```

```python
sns.regplot(x= df['total_bill'], y = df['size'])
```

JOIN PLOT: ITS THE COMBINATION OF REGPLOT AND  SCATTER PLOT and histogram showing the distribution

```python
sns.jointplot(x = df['total_bill'], y=df['tip'])
```

PAIR PLOT: IT PAIRS ONLY THE NUMERICAL COLUMN

```python
sns.pairplot(df)
```

countplot

```python
#its for categorical data
sns.countplot(x=df['size'])
```

```python
sns.countplot(x=df['size'],hue=df['time'])
```

BAR PLOT: IT TAKES IN ONE CATEGORICAL DATA AND ONE CONTINOUS DATA

```python
sns.barplot(x = df ['time'], y = df['tip'])
```

```python
sns.barplot(x = df ['time'], y = df['tip'], estimator=max)
```

```python
sns.barplot(x = df ['time'], y = df['tip'], estimator=max, hue=df['sex'])
```

```python
#you change color using palette
sns.barplot(x = df ['time'], y = df['tip'], estimator=max,hue=df['sex'],palette='gist_heat_r')
```

```python
sns.violinplot(x = df ['sex'], y = df['total_bill'], estimator= max, hue=df['time'],palette='gist_heat_r')
```

```python
sns.violinplot(x = df ['sex'], y = df['total_bill'], estimator= max, hue=df['time'],palette='gist_heat_r',split=True)
```

```python
sns.swarmplot(x = df ['sex'], y = df['total_bill'],  hue=df['time'],palette='gist_heat_r')
```

```python
sns.stripplot(x = df ['sex'], y = df['total_bill'],  hue=df['time'],palette='gist_heat_r')
```

```python
#BOX PLOT TO DETECT OUTLAYER
#the black arrows represent the outlayer
#outlayer means the values that are far above the other values
sns.boxplot(y = df['total_bill'])
```

```python
flight = sns.load_dataset('flights')
df = pd.pivot_table(flight,values = 'passengers',index = 'year', columns = 'month')
df.head()
```

#heatmap is the colorful representation of data , the thicker the data the lower the value the lighter the color the higher the value(HEAT PLOT)

```python
#heatmap is the colorful representation of data , the thicker the data the lower the value the lighter the color the higher the value
plt.figure(figsize=(15,8))
sns.heatmap(df,annot = True, fmt='0.3g')
```

CLUSTER MAP(IT CLUSTERS ANY YEAR THAT HAS DXM VALUES TOGETHER)

```python
plt.figure(figsize=(15,8))
sns.clustermap(df,annot = True, fmt='0.3g')
```

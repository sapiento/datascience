```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import sklearn
```

```python
data = pd.read_csv('sample_pivot.csv')
```

```python
data
```

```python
data.describe()
```

```python
x1 = data['Sales']
y = data['Units']
```

```python
plt.scatter(x1,y)
plt.xlabel('Sales',fontsize=20)
plt.ylabel('Units',fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
sns.set()
plt.scatter(x1,y)
yhat=0*x1 + 0
fig = plt.plot(x1,yhat, lw=4, c='green', label='regression line')
plt.xlabel('Units',fontsize=20)
plt.ylabel('Sales',fontsize=20)
plt.show()
```

```python
sns.set()
plt.scatter(x1,y)
yhat=0*x1 + 0
fig = plt.plot(x1,yhat, lw=4, c='green', label='regression line')
plt.xlabel('Units',fontsize=20)
plt.ylabel('Sales',fontsize=20)

plt.xlim(0)
plt.ylim(0)
plt.show()
```
